```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Initialize Independent Variables (X)
Nclass = 1000
X1 = np.random.randn(Nclass, 2) + np.array([0,-2])
X2 = np.random.randn(Nclass, 2) + np.array([2,2])
X3 = np.random.randn(Nclass, 2) + np.array([-2,2])
X = np.vstack([X1,X2,X3])
X
```

```python
# Initialize Dependent Variable (y)
y = np.array([0]*Nclass + [1]*Nclass + [2]*Nclass)
y
```

```python
# Plot Scatter
plt.scatter(X[:,0], X[:,1], c = y, s = 100, alpha = 0.3)
plt.show()
```

```python
# Initialize Weight & Bias
#input
D = 2
#hidden
M = 3
#output
K = 3
W1 = np.random.randn(D,M)
b1 = np.random.randn(M)
W2 = np.random.randn(M,K)
b2 = np.random.randn(K)
b2
```

```python
# Feedforward Pass

## Sigmoid Activation Function pada Hidden Layer
hidden_layer = 1 / (1+np.exp(-X.dot(W1) - b1))
```

```python
## Softmax Activation Function pada Output Layer
A = hidden_layer.dot(W2) + b2
expA = np.exp(A)
output_layer = expA / expA.sum(axis = 1, keepdims = True)
P = np.argmax ( output_layer, axis = 1)
```

```python
## Classification Rate
n_correct = 0
n_total = 0

for i in range(len(y)):
    n_total += 1
    if y[i] == P[i]:
        n_correct += 1
        rate = (float (n_correct) / n_total) * 100
print("Classification rate dengan pengalaman (weight & bias) acak menghasilkan:", rate,"% Ketepatan")
```

```python
from numpy import concatenate, sort

result = concatenate((y,P))
result.sort(kind='mergesort')
for n, v in zip(y, P):
    print("{} , Predicted Value=> {}".format(n, v))
```
